###### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('SINASC.csv', sep=',')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df2 = df[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
df2.dtypes
```

```python
df2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,6))
plt.tight_layout()

sns.countplot(ax = axes[0], x='APGAR1', data=df2)
sns.countplot(ax = axes[1], x='APGAR5', data=df2)
sns.countplot(ax = axes[2], x='ESTCIVMAE', data=df2)
plt.subplots_adjust(wspace=0.2, hspace=0)

fig2, axes2 = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
plt.tight_layout()

sns.countplot(ax = axes2[0], x='SEXO', data=df2)
sns.countplot(ax = axes2[1], x='GESTACAO', data=df2).tick_params(axis='x', labelrotation=70)
plt.subplots_adjust(wspace=0.2, hspace=0)
```
